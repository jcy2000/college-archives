```python
import csv

results1 = []
results2 = []
results3 = []
    

with open("/Users/chris/Desktop/mapData/raceA.csv",newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',',quotechar='|' )
    for row in reader:
     
        lat = float(row[0])
        long = float(row[1])
        temp = [lat,long]
        if(int(row[4])==1):
            results1.append(temp)
        elif(int(row[4])==2):
            results2.append(temp)
        elif(int(row[4])==3):
            results3.append(temp)
        
        
        

    
A1cords = results1
A2cords = results2
A3cords = results3
m=folium.Map(location=[42.571688,-88.522861], zoom_start=14)
f1=folium.FeatureGroup("A1")
f2=folium.FeatureGroup("A2")
f3=folium.FeatureGroup("A3")
line_1=folium.vector_layers.PolyLine(A1cords,popup='<b>Path of boat A1</b>',tooltip='A1',color='blue',weight=1).add_to(f1)
line_2=folium.vector_layers.PolyLine(A2cords,popup='<b>Path of boat A2</b>',tooltip='A2',color='green',weight=1).add_to(f2)
line_3=folium.vector_layers.PolyLine(A3cords,popup='<b>Path of boat A3</b>',tooltip='A3',color='red',weight=1).add_to(f3)


f1.add_to(m)
f2.add_to(m)
f3.add_to(m)
folium.LayerControl().add_to(m)
m
m.save('/Users/chris/Desktop/mapData/AraceMap.html')

```
